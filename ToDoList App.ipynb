```python
todoList = []
user_input=""

def add_task() :
    input_add=""
    while input_add != "q":
        input_add = input("Enter a new task or Enter 'q'to quit add")
        if input_add != "q":
            todoList.append(input_add)
    

def del_task():
    input_del=""
    while input_del != "q":
        input_del = input("Enter a task to delete or Enter 'q'to quit")
        if input_del != "q":
            if input_del in todoList: 
                todoList.remove(input_del) 
    
def print_all():
    for task in todoList:
        print (task)

def del_all_task():
    todoList.clear()
    
while user_input != "q":
    user_input = input("Enter 'a' to Add your task \nEnter 'd' to delete a task \nEnter 'p' to Print List \nEnter 'da' to delete all task \nEnter 'q' to Quit ")

    if user_input == "a":
        add_task()
    elif user_input == "d":
        del_task()
    elif user_input == "p":
        print_all()
    elif user_input == "da":
        del_all_task()
    
```
